```
using System.IO;

var filedata = File.ReadAllLines(@"..\day13testsl.input");

List<(int,int)> points = new();

List<(int,int)> Foldx (List<(int,int)> points, int x)
{
    var keep = points.Where(p=>p.Item1<=x);
    var transform = points.Where(p=>p.Item1>x).Select(p=>(x*2-p.Item1, p.Item2));
    keep = keep.Union(transform);
    return keep.ToList();
}
List<(int,int)> Foldy (List<(int,int)> points, int y)
{
    var keep = points.Where(p=>p.Item2<=y);
    var transform = points.Where(p=>p.Item2>y).Select(p=>(p.Item1, y*2-p.Item2));
    keep = keep.Union(transform);
    return keep.ToList();
}

void savemap (List<(int,int)> points, string filename, string message)
{
    var mx = points.Max(p=>p.Item1)+1;
    var my = points.Max(p=>p.Item2)+1;
    
    List<string> wr = new() {message};
    for (var j=0;j<my;j++)
    {
        StringBuilder line = new();
        for (var i=0;i<mx;i++)
        {
            line.Append(points.Any(p=>p.Item1==i && p.Item2==j)?"#":".");
        }
        wr.Add(line.ToString());
    }        
    wr.Add("");
    File.WriteAllLines(filename, wr);
}
void printmap (List<(int,int)> points)
{
    var mx = points.Max(p=>p.Item1)+1;
    var my = points.Max(p=>p.Item2)+1;
    
    for (var j=0;j<my;j++)
    {
        for (var i=0;i<mx;i++)
        {
            Console.Write($"{(points.Any(p=>p.Item1==i && p.Item2==j)?'#':'.')}");
        }
        Console.WriteLine();
    }
        
}

bool inputdone = false;

var i=0;
foreach (var line in filedata)
{
    if (string.IsNullOrEmpty(line))
    {
        inputdone=true;
        Console.WriteLine($"There are {points.Count()} points");
        //savemap(points, $"test{i++}.txt", $"There are {points.Count()} points");
        //printmap(points);
        continue;
    }
    if (!inputdone)
    {
        var p = line.Split(',').Select(Int32.Parse).ToArray();
        points.Add((p[0],p[1]));
    }
    else
    {
        if (line.StartsWith("fold along y="))
            points = Foldy(points, Int32.Parse(line.Substring(13)));
        else if (line.StartsWith("fold along x="))
            points = Foldx(points, Int32.Parse(line.Substring(13)));
        Console.WriteLine($"{line} results in {points.Count()} points");
        //savemap(points, $"test{i++}.txt", $"{line} results in {points.Count()} points");
        //printmap(points);
    }
}
printmap(points);
```
